```python
#define : converting datatype :in sales table : date to date
#code 
sales['date']=pd.to_datetime(sales['date'])
```

```python
#test
sales.info()
```

```python
#define : converting datatype :in stores table: Store_Open_Date to date
#code 
stores['store_open_date']=pd.to_datetime(stores['store_open_date'])
#test 
stores.info()
```

```python
#define : converting datatype :in stores table: Store_Location to categorical & product_category to categorical from product table
#code 
stores['store_location']=stores['store_location'].astype('category')
product['product_category'] = product['product_category'].astype('category')

#test 
stores.info()
product.info()
```

```python
#define :  in product table : product_price and product_cost to float 
product['product_price'] = product['product_price'].str.replace('$', '').astype(float)
product['product_cost'] = product['product_cost'].str.replace('$', '').astype(float)
```

```python
#test 
product.info()
product.head()
product.describe()
```

```python
#define cleaning 3 : add a profit  column 
# code
product["profit"]=product["product_price"] - product["product_cost"]
#test
product
```

```python
stores.columns
```

```python
#define : make a year/month/quarter column from the table 
#code
sales['year'] = sales['date'].dt.year
sales['quarter'] = sales['date'].dt.quarter
sales['month'] = sales['date'].dt.month
#test
sales
```

```python
sales.info()
```

```python
#define create a column to know years of operation of each store
#code
stores['year'] = stores['store_open_date'].dt.year
stores['year_operation'] = stores['year'].max()-stores['year']
#drop year column because we don't need it in analysis 
stores.drop(['year'],axis=1,inplace=True)
```

```python
#test
stores.head()
```

```python
#define add inventory id column in the inverntry table
#code
x=np.arange(0,len(inventory))
inventory["inventory_id"] = x
inventory = inventory[["inventory_id","store_id","product_id","stock_on_hand"]]
```

```python
inventory.head()
```

```python
#define : merge table 
#code 
salesXproducts=pd.merge(sales,product, on="product_id")
salesXstores=pd.merge(sales,stores, on="store_id")
inventoryXproducts=pd.merge(inventory, product, on="product_id")
inventoryXstores=pd.merge(inventory, stores, on="store_id")
```
